```python
import pandas as pd
import numpy as np
import warnings
warnings.simplefilter("ignore")

df = pd.read_csv('Group_20_data_cleaned.csv')
df.drop('Unnamed: 0', axis = 1 , inplace = True)
df["Letter"] = pd.factorize(df['Letter'])[0]
```

## linear dimensionality reductions - PCA 

```python
import matplotlib.pyplot as plt
import numpy as np

X, y = df.iloc[:,1:].to_numpy(), df.Letter.to_numpy()
```

```python
import numpy as np
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X)

linear_reduced_data = pca.transform(X)

plt.scatter(linear_reduced_data.take((0), axis=1), 
            linear_reduced_data.take((1), axis=1), 
            c = y, 
            alpha=0.3)
plt.show()
```

```python
A_and_B = np.array([True if i==0 or i==1 else False for i in y ])

plt.scatter(linear_reduced_data[A_and_B].take((0), axis=1), 
            linear_reduced_data[A_and_B].take((1), axis=1), 
            c = y[A_and_B], 
            alpha=0.3)
plt.show()
```

```python
import numpy as np
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X)

linear_reduced_data = pca.transform(X)

plt.scatter(linear_reduced_data.take((0), axis=1), 
            linear_reduced_data.take((1), axis=1), 
            c = y, 
            alpha=0.3)
plt.show()
```

## LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.datasets import make_multilabel_classification
# This produces a feature matrix of token counts, similar to what
# CountVectorizer would produce on text.

lda = LinearDiscriminantAnalysis(n_components=2)
lda.fit(X, y)
lda_reduced_data = lda.transform(X)


plt.scatter(lda_reduced_data.take((0), axis=1), 
            lda_reduced_data.take((1), axis=1), 
            c = y, 
            alpha=0.3)
plt.show()
```

```python
A_and_B = np.array([True if i==0 or i==1 else False for i in y ])

plt.scatter(lda_reduced_data[A_and_B].take((0), axis=1), 
            lda_reduced_data[A_and_B].take((1), axis=1), 
            c = y[A_and_B], 
            alpha=0.3)
plt.show()
```
